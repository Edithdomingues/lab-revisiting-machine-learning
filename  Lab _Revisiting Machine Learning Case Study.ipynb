```python
columns_above_threshold['column_name'].unique()
```

#### Remove those columns from the dataframe

```python
def drop_columns(df, drop_list):
    for i in drop_list:
        df = df.drop(i, axis = 1)
    return df
```

```python
drop_columns = drop_columns(learning_df, ['RDATE_5', 'RAMNT_5', 'RDATE_3', 'RAMNT_3', 'RDATE_4', 'RAMNT_4',
       'RAMNT_6', 'RDATE_6', 'RAMNT_15', 'RDATE_15', 'RDATE_23',
       'RAMNT_23', 'RDATE_20', 'RAMNT_20', 'RAMNT_7', 'RDATE_7',
       'RAMNT_17', 'RDATE_17', 'RDATE_21', 'RAMNT_21', 'RAMNT_10',
       'RDATE_10', 'RAMNT_13', 'RDATE_13', 'NUMCHLD'])
```

#### Reduce the number of categories in the column GENDER. The column should only have either "M" for males, "F" for females, and "other" for all the rest


#### Note that there are a few null values in the column. We will first replace those null values using the code below:

```python
print(learning_df['GENDER'].value_counts())
learning_df['GENDER'] = learning_df['GENDER'].fillna('F')
```

#### Check if filled out correctly

```python
learning_df['GENDER'].value_counts()
```

#### Check for space

```python
learning_df['GENDER'].unique()
```

#### Replace space with 'F'

```python
learning_df['GENDER'] = learning_df['GENDER'].replace(' ', 'F')
learning_df['GENDER'].value_counts()
```

```python
# Separate gender into 3 categories: M, F, other - Using lambda
```

```python
learning_df['GENDER'] = learning_df['GENDER'].apply(lambda x: x if x in ['M', 'F'] else 'other')
learning_df['GENDER'].value_counts()
```
